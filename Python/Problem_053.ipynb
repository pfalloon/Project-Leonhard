# [Project Euler Problem 53](https://projecteuler.net/problem=53)

```python
from math import factorial
def binomial(n, r):
    return factorial(n) // (factorial(r)*factorial(n-r))
```

```python
[binomial(6,r) for r in range(7)]
```

```python
cnt = 0
M = 10**6
for n in range(1,101):
    for r in range(0, n+1):
        if binomial(n,r) > M:
            cnt += 1
cnt
```
